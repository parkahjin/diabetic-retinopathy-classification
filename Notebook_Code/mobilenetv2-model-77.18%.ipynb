```python
# ===================================================================
# Cell 1: ë°ì´í„° ë¡œë“œ + GPU ì„¤ì •
# ===================================================================
import numpy as np
import gc
import tensorflow as tf
import psutil

print("GPU ë©”ëª¨ë¦¬ ë™ì  í• ë‹¹ ì„¤ì • ì¤‘...\n")
gpus = tf.config.list_physical_devices('GPU')
if gpus:
    try:
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
        print("[ì™„ë£Œ] GPU ë©”ëª¨ë¦¬ ë™ì  í• ë‹¹ ì„¤ì • ì™„ë£Œ!\n")
    except RuntimeError as e:
        print(f"[ê²½ê³ ] GPU ì„¤ì • ê²½ê³ : {e}\n")

print("ë°ì´í„° ë¡œë“œ ì¤‘...\n")
DATA_PATH = '/kaggle/input/xy-data'
X_train = np.load(f'{DATA_PATH}/X_train.npy')
y_train = np.load(f'{DATA_PATH}/y_train.npy')
X_val = np.load(f'{DATA_PATH}/X_val.npy')
y_val = np.load(f'{DATA_PATH}/y_val.npy')

print("[ì™„ë£Œ] ë¡œë“œ ì™„ë£Œ!")
print(f"   X_train: {X_train.shape}")
print(f"   y_train: {y_train.shape}")
print(f"   X_val: {X_val.shape}")
print(f"   y_val: {y_val.shape}")

IMG_SIZE = X_train.shape[1]
print(f"\n[í™•ì¸] IMG_SIZE: {IMG_SIZE}")
print(f"   ë°ì´í„° ë©”ëª¨ë¦¬: {(X_train.nbytes + X_val.nbytes) / (1024**3):.2f} GB")

ram = psutil.virtual_memory()
print(f"\n[ë©”ëª¨ë¦¬] RAM: {ram.used/1024**3:.1f}GB / {ram.total/1024**3:.1f}GB ({ram.percent:.1f}%)")

print("\n[ëª©í‘œ] MobileNetV2 ê·¹í•œ ê°œì„ ìœ¼ë¡œ 77.5-78% ë‹¬ì„±!")
print("="*60)
```

```python
# ===================================================================
# Cell 2: ëª¨ë¸ ì •ì˜ (ê²€ì¦ëœ ì„¤ì • + ì•½ê°„ ê°œì„ )
# ===================================================================
print("ğŸ”¥ ê²€ì¦ëœ ì„¤ì •ìœ¼ë¡œ ëª¨ë¸ êµ¬ì¶•...\n")

from tensorflow import keras
from tensorflow.keras import layers, models, callbacks
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.optimizers import Adam

gc.collect()

# ===================================================================
# 1. Focal Loss (ì›ë˜ ì„±ê³µí–ˆë˜ ì„¤ì •!)
# ===================================================================
def focal_loss(gamma=1.5, alpha=0.5):  # ì›ë˜ëŒ€ë¡œ!
    def focal_loss_fixed(y_true, y_pred):
        epsilon = tf.keras.backend.epsilon()
        y_pred = tf.clip_by_value(y_pred, epsilon, 1.0 - epsilon)
        
        y_true = tf.cast(y_true, tf.float32)
        p_t = tf.where(tf.equal(y_true, 1), y_pred, 1 - y_pred)
        alpha_t = tf.where(tf.equal(y_true, 1), alpha, 1 - alpha)
        focal_weight = tf.pow(1.0 - p_t, gamma)
        focal_loss = -alpha_t * focal_weight * tf.math.log(p_t)
        
        return tf.reduce_mean(focal_loss)
    
    return focal_loss_fixed

print("[ì„¤ì •] Focal Loss: gamma=1.5 (ê²€ì¦ë¨), alpha=0.5\n")

# ===================================================================
# 2. ë°ì´í„° ì¦ê°• (ì›ë˜ ì„¤ì • + ì•½ê°„ë§Œ ì¶”ê°€)
# ===================================================================
data_augmentation = keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.1),    # ì›ë˜ëŒ€ë¡œ
    layers.RandomZoom(0.1),        # ì›ë˜ëŒ€ë¡œ
    layers.RandomContrast(0.1),    # ì´ê²ƒë§Œ ì¶”ê°€
], name='data_augmentation')

print("[ì„¤ì •] ë°ì´í„° ì¦ê°• (ê²€ì¦ëœ ì„¤ì •):")
print("      - RandomFlip")
print("      - RandomRotation: 0.1")
print("      - RandomZoom: 0.1")
print("      - RandomContrast: 0.1 (ì¶”ê°€)\n")

# ===================================================================
# 3. MobileNetV2
# ===================================================================
print("MobileNetV2 ë¡œë”©...\n")

base_model = MobileNetV2(
    input_shape=(IMG_SIZE, IMG_SIZE, 3),
    include_top=False,
    weights='imagenet'
)

print(f"[ì™„ë£Œ] MobileNetV2 ë¡œë“œ!")
print(f"      ì „ì²´ layers: {len(base_model.layers)}\n")

# ===================================================================
# 4. Fine-tuning (ì›ë˜ ì„¤ì • + ì•½ê°„ë§Œ ì¦ê°€)
# ===================================================================
FINE_TUNE_LAYERS = 50  # ì›ë˜ëŠ” 40, 50ìœ¼ë¡œ ì•½ê°„ë§Œ ì¦ê°€

print(f"[ì„¤ì •] Fine-tuning:")
print(f"      - Freeze: ì²˜ìŒ {len(base_model.layers) - FINE_TUNE_LAYERS} layers")
print(f"      - Fine-tune: ë§ˆì§€ë§‰ {FINE_TUNE_LAYERS} layers\n")

for layer in base_model.layers[:-FINE_TUNE_LAYERS]:
    layer.trainable = False
    
for layer in base_model.layers[-FINE_TUNE_LAYERS:]:
    layer.trainable = True

# ===================================================================
# 5. ëª¨ë¸ êµ¬ì¶• (ì›ë˜ ì„¤ì • ê·¸ëŒ€ë¡œ)
# ===================================================================
model = models.Sequential([
    layers.Input(shape=(IMG_SIZE, IMG_SIZE, 3)),
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.BatchNormalization(),
    layers.Dropout(0.5),
    layers.Dense(256, activation='relu'),
    layers.BatchNormalization(),
    layers.Dropout(0.3),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.2),
    layers.Dense(1, activation='sigmoid', dtype='float32')
])

print("[ì™„ë£Œ] ëª¨ë¸ êµ¬ì¶• (ê²€ì¦ëœ êµ¬ì¡°)\n")

# ===================================================================
# 6. ì»´íŒŒì¼ (ì›ë˜ ì„¤ì • ê·¸ëŒ€ë¡œ)
# ===================================================================
model.compile(
    optimizer=Adam(learning_rate=0.0005),  # ì›ë˜ëŒ€ë¡œ
    loss=focal_loss(gamma=1.5, alpha=0.5),
    metrics=['accuracy', 
             tf.keras.metrics.AUC(name='auc'),
             tf.keras.metrics.Precision(name='precision'),
             tf.keras.metrics.Recall(name='recall')]
)

print("="*60)
print("ğŸ“Š ëª¨ë¸ ìš”ì•½")
print("="*60)
model.summary()

print("\n" + "="*60)
print("âœ… ê²€ì¦ëœ ëª¨ë¸ ì¤€ë¹„ ì™„ë£Œ!")
print("="*60)
print(f"ëª¨ë¸: MobileNetV2")
print(f"ì†ì‹¤: Focal Loss (gamma=1.5)")
print(f"LR: 0.0005")
print(f"Fine-tuning: {FINE_TUNE_LAYERS} layers")
print(f"ì¦ê°•: 4ê°œ (Contrast ì¶”ê°€)")
print("="*60)
```

```python
# ===================================================================
# Cell 3: í•™ìŠµ (ê²€ì¦ëœ ì„¤ì • + Epochsë§Œ ì¦ê°€)
# ===================================================================
print("\nğŸš€ í•™ìŠµ ì‹œì‘!")
print("="*60)

from sklearn.utils import class_weight
import pickle

gc.collect()

class_weights = class_weight.compute_class_weight(
    'balanced',
    classes=np.unique(y_train),
    y=y_train
)
class_weight_dict = {i: class_weights[i] for i in range(len(class_weights))}
print(f"âœ… Class weights: {class_weight_dict}\n")

# Dataset ìƒì„±
print("Dataset ìƒì„±...\n")
BATCH_SIZE = 10

def augment_data(image, label):
    image = data_augmentation(image, training=True)
    return image, label

train_dataset = tf.data.Dataset.from_tensor_slices((X_train, y_train))
train_dataset = train_dataset.shuffle(buffer_size=10000, seed=42)
train_dataset = train_dataset.batch(BATCH_SIZE)
train_dataset = train_dataset.map(augment_data, num_parallel_calls=tf.data.AUTOTUNE)
train_dataset = train_dataset.prefetch(tf.data.AUTOTUNE)

val_dataset = tf.data.Dataset.from_tensor_slices((X_val, y_val))
val_dataset = val_dataset.batch(BATCH_SIZE)
val_dataset = val_dataset.prefetch(tf.data.AUTOTUNE)

print(f"[ì™„ë£Œ] Dataset ì¤€ë¹„!")
print(f"   Train: {len(y_train):,} samples")
print(f"   Val: {len(y_val):,} samples\n")

del X_train
gc.collect()
print("[ì™„ë£Œ] ë©”ëª¨ë¦¬ ì •ë¦¬\n")

# Callbacks
checkpoint = callbacks.ModelCheckpoint(
    'best_model_improved.weights.h5',
    monitor='val_accuracy',
    save_best_only=True,
    save_weights_only=True,
    mode='max',
    verbose=1
)

early_stop = callbacks.EarlyStopping(
    monitor='val_loss',
    patience=10,  # ì›ë˜ëŒ€ë¡œ
    restore_best_weights=True,
    verbose=1
)

reduce_lr = callbacks.ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=3,  # ì›ë˜ëŒ€ë¡œ
    min_lr=1e-7,
    verbose=1
)

EPOCHS = 35  # 25 â†’ 35 (ì´ê²ƒë§Œ ì¦ê°€)

print("="*60)
print("ğŸ”¥ í•™ìŠµ ì„¤ì •")
print("="*60)
print(f"Epochs: {EPOCHS} (25â†’35 ì¦ê°€)")
print(f"Batch: {BATCH_SIZE}")
print(f"LR: 0.0005")
print(f"Focal Loss: gamma=1.5 (ê²€ì¦ë¨)")
print(f"ëª©í‘œ: 76.5-77.5%")
print("="*60)

try:
    history = model.fit(
        train_dataset,
        epochs=EPOCHS,
        validation_data=val_dataset,
        class_weight=class_weight_dict,
        callbacks=[checkpoint, early_stop, reduce_lr],
        verbose=1
    )
    
    print("\n" + "="*60)
    print("âœ… í•™ìŠµ ì™„ë£Œ!")
    print("="*60)
    
    best_val_acc = max(history.history['val_accuracy'])
    best_epoch = history.history['val_accuracy'].index(best_val_acc) + 1
    
    print(f"\nğŸ“Š ìµœì¢… ê²°ê³¼:")
    print(f"   ìµœê³  ì •í™•ë„: {best_val_acc*100:.2f}%")
    print(f"   ë‹¬ì„± Epoch: {best_epoch}/{EPOCHS}")
    
    improvement = (best_val_acc - 0.7635) * 100
    print(f"\nğŸ¯ ê°œì„ :")
    print(f"   ì´ì „: 76.35%")
    print(f"   í˜„ì¬: {best_val_acc*100:.2f}%")
    print(f"   í–¥ìƒ: {improvement:+.2f}%p")
    
    if best_val_acc >= 0.77:
        print(f"\nğŸ‰ ëª©í‘œ ë‹¬ì„±!")
    elif best_val_acc >= 0.765:
        print(f"\nâœ… ê°œì„  ì„±ê³µ!")
    else:
        print(f"\nğŸ“Š ì•½ê°„ ê°œì„ ")
    
    # ===================================================================
    # ì €ì¥ (ê²€ì¦ ì¶”ê°€!) â­â­â­
    # ===================================================================
    print("\n" + "="*60)
    print("ğŸ’¾ ì €ì¥ í”„ë¡œì„¸ìŠ¤ ì‹œì‘ (ê²€ì¦ í¬í•¨)")
    print("="*60)
    
    # 1. history ì €ì¥
    print("\nğŸ“ Step 1: History ì €ì¥...")
    with open('history_improved.pkl', 'wb') as f:
        pickle.dump(history.history, f)
    print("   âœ… history_improved.pkl ì €ì¥ ì™„ë£Œ")
    
    # 2. í˜„ì¬ ëª¨ë¸ ê²€ì¦ (EarlyStoppingì´ best weights ë³µì›í–ˆëŠ”ì§€ í™•ì¸)
    print("\nğŸ” Step 2: í˜„ì¬ ëª¨ë¸ ê²€ì¦...")
    current_loss, current_acc = model.evaluate(val_dataset, verbose=0)
    print(f"   í˜„ì¬ ëª¨ë¸ ì •í™•ë„: {current_acc*100:.2f}%")
    
    # 3. ì„±ëŠ¥ ì²´í¬
    if current_acc >= 0.75:  # 75% ì´ìƒì¼ ë•Œë§Œ ì €ì¥
        print(f"   âœ… ì„±ëŠ¥ OK ({current_acc*100:.2f}% >= 75%)")
        
        # 4. .keras íŒŒì¼ ì €ì¥
        print("\nğŸ’¾ Step 3: ëª¨ë¸ ì €ì¥...")
        model.save('final_mobilenet_improved.keras')
        print("   âœ… final_mobilenet_improved.keras ì €ì¥ ì™„ë£Œ")
        
        # 5. ì €ì¥ ê²€ì¦ (ì¬ë¡œë“œí•´ì„œ í™•ì¸)
        print("\nğŸ” Step 4: ì €ì¥ ê²€ì¦ (ì¬ë¡œë“œ í…ŒìŠ¤íŠ¸)...")
        try:
            model_test = tf.keras.models.load_model(
                'final_mobilenet_improved.keras',
                custom_objects={'focal_loss_fixed': focal_loss(gamma=1.5, alpha=0.5)}
            )
            reload_loss, reload_acc = model_test.evaluate(val_dataset, verbose=0)
            print(f"   ì¬ë¡œë“œ ëª¨ë¸ ì •í™•ë„: {reload_acc*100:.2f}%")
            
            # ì •í™•ë„ ì°¨ì´ í™•ì¸
            acc_diff = abs(reload_acc - current_acc) * 100
            if acc_diff < 0.5:  # 0.5%p ì´ë‚´ ì°¨ì´ë©´ OK
                print(f"   âœ… ê²€ì¦ í†µê³¼! (ì°¨ì´: {acc_diff:.2f}%p)")
                print("\n" + "="*60)
                print("ğŸ‰ ì €ì¥ ì„±ê³µ!")
                print("="*60)
                print("âœ… ì €ì¥ëœ íŒŒì¼:")
                print("   - history_improved.pkl")
                print("   - final_mobilenet_improved.keras (ê²€ì¦ë¨!)")
                print("   - best_model_improved.weights.h5")
            else:
                print(f"   âš ï¸ ê²½ê³ : ì •í™•ë„ ì°¨ì´ í¼ ({acc_diff:.2f}%p)")
                print("   í•˜ì§€ë§Œ ì €ì¥ì€ ì™„ë£Œë¨")
                
        except Exception as e:
            print(f"   âš ï¸ ì¬ë¡œë“œ ê²€ì¦ ì‹¤íŒ¨: {e}")
            print("   í•˜ì§€ë§Œ ì €ì¥ì€ ì™„ë£Œë¨")
        
        import os
        size = os.path.getsize('final_mobilenet_improved.keras') / (1024**2)
        print(f"\nğŸ“¦ íŒŒì¼ í¬ê¸°: {size:.1f} MB")
        
    else:
        print(f"   âŒ ì„±ëŠ¥ ë¶€ì¡± ({current_acc*100:.2f}% < 75%)")
        print("   .keras íŒŒì¼ ì €ì¥ ì•ˆ í•¨ (weightsë§Œ ìˆìŒ)")
    
    print("\n" + "="*60)
    print("ğŸš¨ Save Version í•˜ì„¸ìš”!")
    print("="*60)
    print("Version name: 'phase1-mobilenet-verified'")
    print("="*60)
    
except Exception as e:
    print(f"\nâŒ ERROR: {e}")
    import traceback
    traceback.print_exc()
```

```python
# ===================================================================
# 76.64% ëª¨ë¸ ë¡œë“œ ë° ê²€ì¦
# ===================================================================
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.optimizers import Adam
import numpy as np
from sklearn.metrics import accuracy_score
import gc

gc.collect()

print("="*60)
print("ğŸ”§ 76.64% ëª¨ë¸ ë¡œë“œ")
print("="*60)

# Focal Loss
def focal_loss(gamma=1.5, alpha=0.5):
    def focal_loss_fixed(y_true, y_pred):
        epsilon = tf.keras.backend.epsilon()
        y_pred = tf.clip_by_value(y_pred, epsilon, 1.0 - epsilon)
        
        y_true = tf.cast(y_true, tf.float32)
        p_t = tf.where(tf.equal(y_true, 1), y_pred, 1 - y_pred)
        alpha_t = tf.where(tf.equal(y_true, 1), alpha, 1 - alpha)
        focal_weight = tf.pow(1.0 - p_t, gamma)
        focal_loss = -alpha_t * focal_weight * tf.math.log(p_t)
        
        return tf.reduce_mean(focal_loss)
    
    return focal_loss_fixed

# ëª¨ë¸ êµ¬ì¡° ì¬ìƒì„± (Phase 1ê³¼ ë™ì¼)
IMG_SIZE = 96
FINE_TUNE_LAYERS = 50  # 76.64% ëª¨ë¸ ì„¤ì •

print("\nğŸ”§ ëª¨ë¸ êµ¬ì¡° ì¬ìƒì„± ì¤‘...")

base_model = MobileNetV2(
    input_shape=(IMG_SIZE, IMG_SIZE, 3),
    include_top=False,
    weights='imagenet'
)

for layer in base_model.layers[:-FINE_TUNE_LAYERS]:
    layer.trainable = False
    
for layer in base_model.layers[-FINE_TUNE_LAYERS:]:
    layer.trainable = True

model = models.Sequential([
    layers.Input(shape=(IMG_SIZE, IMG_SIZE, 3)),
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.BatchNormalization(),
    layers.Dropout(0.5),
    layers.Dense(256, activation='relu'),
    layers.BatchNormalization(),
    layers.Dropout(0.3),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.2),
    layers.Dense(1, activation='sigmoid', dtype='float32')
])

model.compile(
    optimizer=Adam(learning_rate=0.0005),
    loss=focal_loss(gamma=1.5, alpha=0.5),
    metrics=['accuracy']
)

print("âœ… ëª¨ë¸ êµ¬ì¡° ìƒì„± ì™„ë£Œ")

# 76.64% weights ë¡œë“œ
print("\nğŸ“¥ 76.64% weights ë¡œë“œ ì¤‘...")
model.load_weights('/kaggle/input/phase1-mobilenet-76-64pct/best_model_improved_76.weights.h5')
print("âœ… weights ë¡œë“œ ì™„ë£Œ!")

# ê²€ì¦ ë°ì´í„° ë¡œë“œ
print("\nğŸ“¥ ê²€ì¦ ë°ì´í„° ë¡œë“œ ì¤‘...")
X_val = np.load('/kaggle/input/xy-data/X_val.npy')
y_val = np.load('/kaggle/input/xy-data/y_val.npy')

print(f"âœ… X_val shape: {X_val.shape}")
print(f"âœ… y_val shape: {y_val.shape}")

# ëª¨ë¸ ê²€ì¦ (0-255 ë²”ìœ„ ê·¸ëŒ€ë¡œ!)
print("\nğŸ” ëª¨ë¸ ì‘ë™ í™•ì¸ (ì •ê·œí™” ì—†ì´)")
print("="*60)

X_val_test = X_val[:100]  # ì •ê·œí™” ì—†ìŒ!
pred_test = model.predict(X_val_test, batch_size=32, verbose=0)

print(f"\nì˜ˆì¸¡ ìƒ˜í”Œ (ì²˜ìŒ 10ê°œ):")
print(pred_test[:10].flatten())
print(f"\nì˜ˆì¸¡ í†µê³„:")
print(f"  Min:  {pred_test.min():.4f}")
print(f"  Max:  {pred_test.max():.4f}")
print(f"  Mean: {pred_test.mean():.4f}")
print(f"  Std:  {pred_test.std():.4f}")

# ì „ì²´ ê²€ì¦
pred_all = model.predict(X_val, batch_size=32, verbose=0)
y_pred = (pred_all > 0.5).astype(int).flatten()
acc = accuracy_score(y_val, y_pred)

print(f"\nğŸ“Š ê²€ì¦ ì •í™•ë„: {acc*100:.2f}%")

print("\n" + "="*60)
if pred_test.std() > 0.01 and acc > 0.75:
    print(f"âœ… 76.64% ëª¨ë¸ ë³µì› ì„±ê³µ!")
    print("ë‹¤ìŒ ë‹¨ê³„: TTA ì§„í–‰")
else:
    print("âŒ ë¬¸ì œ ë°œìƒ")
print("="*60)
```

```python
# ===================================================================
# TTA (Test Time Augmentation)
# ===================================================================
print("="*60)
print("ğŸ”„ TTA ì‹œì‘ (76.64% ëª¨ë¸)")
print("="*60)

from sklearn.metrics import accuracy_score
import numpy as np

# X_val ê·¸ëŒ€ë¡œ ì‚¬ìš© (0-255 ë²”ìœ„)
X_val_input = X_val

print("\nğŸ”„ TTA ì¦ê°• ë²„ì „ë³„ ì˜ˆì¸¡ ì¤‘...")

# 1. ì›ë³¸
print("   1/5: ì›ë³¸")
pred_original = model.predict(X_val_input, batch_size=32, verbose=0)

# 2. ì¢Œìš° ë°˜ì „
print("   2/5: ì¢Œìš° ë°˜ì „")
pred_flip_lr = model.predict(np.flip(X_val_input, axis=2), batch_size=32, verbose=0)

# 3. ìƒí•˜ ë°˜ì „
print("   3/5: ìƒí•˜ ë°˜ì „")
pred_flip_ud = model.predict(np.flip(X_val_input, axis=1), batch_size=32, verbose=0)

# 4. ì¢Œìš°+ìƒí•˜
print("   4/5: ì¢Œìš°+ìƒí•˜ ë°˜ì „")
pred_flip_both = model.predict(np.flip(np.flip(X_val_input, axis=1), axis=2), batch_size=32, verbose=0)

# 5. ë°ê¸° ì¡°ì •
print("   5/5: ë°ê¸° ì¡°ì •")
pred_bright = model.predict(np.clip(X_val_input * 1.1, 0, 255), batch_size=32, verbose=0)

# ì•™ìƒë¸”
print("\nğŸ“Š TTA ì•™ìƒë¸” ì¤‘...")
predictions_tta = (pred_original + pred_flip_lr + pred_flip_ud + pred_flip_both + pred_bright) / 5.0

# ê¸°ë³¸ 0.5 threshold
y_pred_tta = (predictions_tta > 0.5).astype(int).flatten()
acc_tta = accuracy_score(y_val, y_pred_tta)

print("\n" + "="*60)
print("âœ… TTA ì™„ë£Œ!")
print("="*60)
print(f"ğŸ“Š Phase 1 ê¸°ë³¸: 76.64%")
print(f"ğŸ”„ TTA ì ìš©:    {acc_tta*100:.2f}%")
print(f"ğŸ“ˆ ê°œì„ :        +{(acc_tta*100 - 76.64):.2f}%p")
print("="*60)
```

```python
# ===================================================================
# Threshold ìµœì í™”
# ===================================================================
print("\n" + "="*60)
print("ğŸ¯ Threshold ìµœì í™”")
print("="*60)

from sklearn.metrics import classification_report, confusion_matrix

best_threshold = 0.5
best_accuracy = 0

print("\nğŸ” Threshold íƒìƒ‰ ì¤‘ (0.30 ~ 0.70)...")
thresholds = np.arange(0.30, 0.71, 0.01)

for threshold in thresholds:
    y_pred = (predictions_tta > threshold).astype(int).flatten()
    accuracy = accuracy_score(y_val, y_pred)
    
    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_threshold = threshold

# ìµœì¢… ì˜ˆì¸¡
y_pred_final = (predictions_tta > best_threshold).astype(int).flatten()

print("\n" + "="*60)
print("âœ… ìµœì í™” ì™„ë£Œ!")
print("="*60)
print(f"ğŸ¯ ìµœì  Threshold: {best_threshold:.3f}")
print(f"ğŸ“Š ìµœì¢… ì •í™•ë„:    {best_accuracy*100:.2f}%")
print("="*60)
print(f"\nğŸ“ˆ ì„±ëŠ¥ ê°œì„  ê³¼ì •:")
print(f"   Phase 1:     76.64%")
print(f"   + TTA:       {acc_tta*100:.2f}% (+{(acc_tta*100-76.64):.2f}%p)")
print(f"   + Threshold: {best_accuracy*100:.2f}% (+{(best_accuracy*100-acc_tta*100):.2f}%p)")
print(f"   â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”")
print(f"   ì´ ê°œì„ :     +{(best_accuracy*100-76.64):.2f}%p")
print("="*60)

print("\nğŸ“Š ìµœì¢… ì„±ëŠ¥ ì§€í‘œ:")
print("="*60)
print(classification_report(y_val, y_pred_final, 
                          target_names=['Normal', 'Abnormal'],
                          digits=4))

print("\ní˜¼ë™ í–‰ë ¬:")
cm = confusion_matrix(y_val, y_pred_final)
print(cm)
print(f"\nTN: {cm[0,0]}, FP: {cm[0,1]}")
print(f"FN: {cm[1,0]}, TP: {cm[1,1]}")
print("="*60)

print("\nğŸ‰ í”„ë¡œì íŠ¸ ì™„ë£Œ!")
print(f"ğŸ“Š ìµœì¢… ì •í™•ë„: {best_accuracy*100:.2f}%")
print("="*60)
```
